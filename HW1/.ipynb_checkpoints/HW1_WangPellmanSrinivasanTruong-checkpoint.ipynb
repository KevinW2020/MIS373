```python
#Rock, Paper, Scissors Game
from pandas import Series, DataFrame
import random
import pandas as pd

#Set up dataframe with all possible combinations, column is computer choice and row is human choice
options = ['Rock','Paper','Scissors']
data = {'Rock':['Tied','Lost','Won'],'Paper':['Won','Tied','Lost'],'Scissors':['Lost','Won','Tied']}
df = DataFrame(data, index = options)

#Set up series keeping track of player's decisions.
count = [0,0,0]
history = Series(count, index = options)

wins = 0
games = 0

#Conditions to whether or not player wants to keep playing
playing = True
while playing == True:
    player_choice = raw_input('Your move: ')
    if (player_choice not in options) and (player_choice != "Quit"):
        player_choice = raw_input('Invalid decision, try again: ')
    elif (player_choice == "Quit"):
        playing = False
        print ("You played "+str(games)+" and won "+str(wins))
    else:
        if history.idxmax() == "Rock":
            comp_choice = "Paper"
        elif history.idxmax() == "Paper":
            comp_choice = "Scissors"
        else:
            comp_choice = "Rock"
        result = df.get_value(comp_choice,player_choice)
        print ("You chose "+player_choice+", the computer chose "+comp_choice+": You "+result)
        if result == "Won":
            wins +=1
        games +=1
        history[player_choice] +=1
```

```python
#Movies.csv (part 2)

#Sets up a dictionary with the actor name as the key and their movies as a list of values
history = {}
mov = open('movies.csv','r')
for line in mov:
    line = line.strip()
    t = line.split(',')
    history.update({t[0]:t[1:len(t)]})

movies = []
#Set up list of all valid movies (unique)
for key, value in history.iteritems():
    for item in value:
        if item not in movies:
            movies.append(item)

#Reads user input and verfies movies are actually in the list
checker = False
while checker == False:
    movie1 = raw_input("First Movie: ")
    if movie1 in movies:
        checker = True
    else:
        print("Invalid choice. Try again: ")
checker = False
while checker == False:
    movie2 = raw_input("Second Movie: ")
    if movie2 in movies:
        checker = True
    else:
        print("Invalid choice. Try again: ")

#Function that does actors in both movies
BothMovies = []
for key, value in history.iteritems():
    if (movie1 in value) and (movie2 in value):
        BothMovies.append(key)
print ("These actors appeared in both "+movie1+" and "+movie2+":"+str(BothMovies))

#Movie1 but not movie2
OneNTwo = []
for key,value in history.iteritems():
    if (movie1 in value) and (movie2 not in value):
        OneNTwo.append(key)
print ("These actors appeared in "+movie1+" but not "+movie2+":"+str(OneNTwo))

#Movie2 but not movie1
TwoNOne = []
for key,value in history.iteritems():
    if (movie1 not in value) and (movie1 in value):
        TwoNOne.append(key)
print ("These actors appeared in "+movie2+" but not "+movie1+":"+str(TwoNOne))
```
